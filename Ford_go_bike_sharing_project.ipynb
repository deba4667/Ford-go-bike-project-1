# **Project Summary -**

This project is based on Exploratory Data Analysis (EDA) of the Ford GoBike dataset for January 2018. The main goal was to understand user behavior, trip patterns, and operational trends without applying machine learning models like regression or classification.

The dataset was cleaned by converting trip durations from seconds to minutes and changing start and end times into datetime formats. Basic checks were performed for missing data and inconsistencies.

During univariate analysis, it was found that the average trip lasted between 10 to 15 minutes. Major stations like San Francisco Ferry Building and Caltrain were among the busiest start and end points.

In bivariate and multivariate analysis, it was observed that subscribers mainly used the bikes during weekdays for shorter, regular commutes. In contrast, customers preferred weekends and had longer average ride durations, reflecting more leisure-oriented use.

Charts were created to visualize trip patterns based on user type and day of the week, which helped clearly highlight these differences.

Key Insights:

	•	Subscribers are more active during weekdays for commuting.
	•	Customers prefer weekends and take longer trips.
	•	Major transport hubs are the busiest bike stations.

Conclusion:

This EDA project provided important operational insights that can help improve bike availability, station planning, and marketing strategies for different user groups. It was completed individually and focused purely on exploratory analysis.

**Data Import and Overview**

```python
# Importing necessary libraries
import pandas as pd

# Loading the dataset
df = pd.read_csv('201801-fordgobike-tripdata.csv')

# Displaying basic information
df.head()
df.info()
df.describe()
```

**Data Cleaning**

Write your steps like this:

	•	Check for missing values.
	•	Drop or fill missing values if needed.
	•	Check for duplicate rows.
	•	Convert date columns (start_time, end_time) to datetime format.

```python
# Checking missing values
df.isnull().sum()

# Dropping rows with missing values
df.dropna(inplace=True)

# Converting date columns to datetime
df['start_time'] = pd.to_datetime(df['start_time'])
df['end_time'] = pd.to_datetime(df['end_time'])
```

**Exploratory Data Analysis (EDA)**

## Solution to the Business Objective

Based on the analysis of Ford GoBike's January 2018 trip data, the following solutions are recommended to achieve business growth:

- *1. Focus on Subscriber Retention:*  
  The majority of users are Subscribers. The company should launch loyalty programs, membership discounts, and special rewards to retain them and encourage frequent usage.

- *2. Target Female and Diverse Users:*  
  Since male users dominate the platform, Ford GoBike should create marketing campaigns and safety initiatives focused on attracting more female and diverse riders.

- *3. Optimize Bike Availability During Peak Hours:*  
  Peak usage is observed at 8 AM and 5 PM (commuting hours). The company should ensure that more bikes are available during these times, especially at popular start stations.

- *4. Age-Specific Promotions:*  
  Most users fall between the age group of 20–40 years. Marketing strategies and ride plans should focus on the needs of young working professionals, such as monthly commuter passes and flexible ride options.

- *5. Improve Station Management:*  
  Certain start and end stations are extremely popular. Ford GoBike should allocate more bikes and docking spaces at these high-demand stations to enhance user experience.

- *6. Weekend Offers for Casual Customers:*  
  Customer rides are higher on weekends. Offering special weekend packages, family ride discounts, or tourist promotions can help convert casual riders into Subscribers.

- *7. Proactive Bike Maintenance:*  
  Bikes that are heavily used should undergo regular maintenance. Tracking usage based on trip duration can help prioritize which bikes need servicing first.

- *8. Implement Dynamic Pricing Strategies:*  
  By adjusting ride prices based on peak and non-peak hours, the company can manage bike demand more efficiently and maximize profits.
